```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = [1, 2, 3, 4]
Y = [5, 6, 7, 8]
Z = [9, 10, 11 ,12]
plt.plot(X,Y,color="red",linestyle="--",linewidth=2,marker="o",label="X-Y")
plt.plot(X,Z,color="blue",linestyle=":",linewidth=2,marker="x",label="X-Z")
plt.xlabel("x-value")
plt.ylabel("y-value")
plt.title("X-Y comparison")
plt.legend()
plt.show()
```

```python
x = ['A', 'B', 'C', 'D']
y = [10, 15, 7, 12]
plt.bar(x, y,color="purple",width=0.6,edgecolor="black",linewidth=1)
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Simple Bar Chart")
for i in range(len(x)):
    plt.text(x[i],y[i]+0.5,y[i],ha="center")
plt.ylim(0,max(y)+2)
plt.grid(True)
plt.show()
```

```python
sizes = [30, 20, 25, 25]
labels = ['A', 'B', 'C', 'D']

plt.pie(sizes,
labels=labels,
autopct='%1.1f%%',
startangle=90, # rotate chart
shadow=True) # adds shadow
plt.show()
```

```python
data = [1, 2, 2, 3, 3, 3, 4, 4, 5, 5, 5, 5]
plt.hist(data, bins=5, color='skyblue', edgecolor='black')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Simple Histogram")
plt.grid(axis='y', linestyle='--',alpha=0.7)
plt.show()
```

```python
data = [7, 8, 5, 6, 9, 7, 8, 10, 4, 6]
plt.boxplot(data)
plt.ylabel("Values")
plt.title("Simple Box Plot")
plt.show()
```

```python
days=["Mon","Tues","Wed","Thurs","Fri"]
cities=["NewYork", "London","Delhi","Tokyo"]

temperature=[
    [22,23,21,24,25],
    [18,19,17,20,21],
    [30,32,31,33,34],
    [25,26,24,27,28]
]
```

```python
fig, axes=plt.subplots(2,2)
city_no=0
for i in range(0,2):
    for j in range(0,2):
        axes[i][j].plot(days,temperature[city_no])
        axes[i][j].set_title(cities[city_no])
        axes[i][j].grid(True)
        city_no=city_no+1

fig.tight_layout()
```
