```python
import numpy as np
import pandas as pd
```

```python
class LinearRegression:
    def __init__(self, learning_rate=0.01, n_iter=1000):
        self.bias=None
        self.weights=M=None
        self.lr=learning_rate
        self.n_iter=n_iter

    def fit(self, X, y): #X_train, y_train
        m,n=X.shape   #(no of samples, no of features)

        #step1-initialising values
        self.bias=0
        self.weights=np.zeros(n)   #will create array of n features

        #Gradient Descent
        for i in range(self.n_iter):
        #step 2 -calculating y predict
            y_pred=self.bias + np.dot(X, self.weights)

            #step 3- calculating error
            db=(1/m) * np.sum(y_pred- y)
            dw=(1/m) * np.dot(X.T,(y_pred-y))

            #step 4-updating weight (convergence theorem)
            self.bias -= self.lr * db
            self.weights -= self.lr * dw
    def predict(self, X):
        y_pred = self.bias + np.dot(X, self.weights)
        return y_pred

                
            
```

```python
X=np.array([[1], [2], [3], [4], [5]])
y=np.array([2, 4, 6, 8, 10])

model = LinearRegression()
model.fit(X,y)

y_pred = model.predict(X)
print(y_pred)
```
